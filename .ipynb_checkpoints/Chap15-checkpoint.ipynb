# Autoencoders

- 基本的には、隠れ層のニューロン数に制限をかけた状態で(Information Bottleneck)、入力と全く同じ出力を得られるようにNeural Networkを訓練するという枠組み


- 普通のautoencoderや、denoising autoencoder, sparse autoencoderなどは、インスタンスの重要な特徴量を覚えられる、とされている


- Autoencoderをうまく使うことで、Neural Network の pre-training を行うことができる


- Latent space を工夫することで、インスタンスを新しく生成することが可能 (variational autoencoder)

※ 発表者が tensorflow より keras に慣れてしまっているため、コードは全て keras で書いています。tensorflow で書く方法は著者の jupyter notebook をご参照ください

```python
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import keras
from keras.layers import Input, Dense
from keras.models import Model
from keras import backend as K
from keras.datasets import mnist

plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
```

## 15.2 非線形性を抜いたautoencoderはPCAを行っている


- 本当？


- PCAの結果と近くなる、というのは間違ってないが、autoencoder は直交性を担保してくれる項がなにも入っていないので、Principal Component Analysis と一致する保証はないように思える


- 3次元空間に作った構造を3→2→3の autoencoder で学習させた場合、latent space がどうなっているかを見てみる

```python
# generating structured data in 3D
import numpy.random as rnd
from sklearn.preprocessing import StandardScaler

rnd.seed(4)
m = 200
w1, w2 = 0.1, 0.3
noise = 0.1

angles = rnd.rand(m) * 3 * np.pi / 2 - 0.5 # random m samples from a uniform distribution [-0.5, 1.5\pi-0.5)
data = np.empty((m, 3))
data[:, 0] = np.cos(angles) + np.sin(angles)/2 + noise * rnd.randn(m) / 2
data[:, 1] = np.sin(angles) * 0.7 + noise * rnd.randn(m) / 2
data[:, 2] = data[:, 0] * w1 + data[:, 1] * w2 + noise * rnd.randn(m)

scaler = StandardScaler() # normalization; 0-mean, unit-variance
X_train = scaler.fit_transform(data[:100])
X_test = scaler.transform(data[100:])
```

```python
from keras.layers import Input, Dense
from keras.models import Model

encoding_dim = 2  # dimension of latent dimension

input_img = Input(shape=(3,)) # input placeholder
encoded = Dense(encoding_dim, activation='linear')(input_img)
decoded = Dense(3, activation='linear')(encoded)

encoder = Model(input_img, encoded) # creating a separate encoder model

autoencoder = Model(input_img, decoded)
autoencoder.compile(optimizer='adam', loss='mse')

autoencoder.fit(X_train, X_train, epochs=1500, batch_size=100,
                shuffle=True, validation_data=(X_test, X_test)) # executed in 6.12s with GTX 1070Ti
```

MNIST autoencoder の latent space がそれぞれどんな特徴量に対応しているかを図示する


→ Gradient descentを利用して、個々のニューロンの出力が最大化されるような input を探す

```python
autoencoder.summary()
```

```python
from keras import backend as K

neuron_index = 0 # 0 to 31 
latent_output = autoencoder.get_layer('dense_3').output
loss = K.mean(latent_output[:,neuron_index])

grads = K.gradients(loss, autoencoder.input)[0]
grads /= (K.sqrt(K.mean(K.square(grads))) + 1e-5) 
# normalizing gradient so that the updates to the input are always in the same range # also to avoide 0 division

# defining function which returns loss and gradients when input is given
iterate = K.function([autoencoder.input], [loss, grads])
```

```python
input_random_data = np.random.random((1, 784)) * 0.3 + 0.5

step = 1.  # this is the magnitude of each gradient update
for i in range(40):
    loss_value, grads_value = iterate([input_random_data])
    # update the input image in the direction that maximizes the loss
    input_random_data += grads_value * step
plt.imshow(input_random_data.reshape(28, 28), cmap='viridis')
```

```python
def plot_several_neurons():
    plt.subplots(4,4,figsize=(16,16))
    for j in range(0,16):
        ax = plt.subplot(4,4,j+1)
        neuron_index = j
        
        latent_output = autoencoder.get_layer('dense_3').output
        loss = K.mean(latent_output[:,neuron_index])

        grads = K.gradients(loss, autoencoder.input)[0]
        grads /= (K.sqrt(K.mean(K.square(grads))) + 1e-5) 
        # normalizing gradient so that the updates to the input are always in the same range # also to avoide 0 division

        # defining function which returns loss and gradients when input is given
        iterate = K.function([autoencoder.input], [loss, grads])
        
        input_random_data = np.random.random((1, 784)) * 0.3 + 0.5

        step = 1.  # this is the magnitude of each gradient update
        for i in range(40):
            loss_value, grads_value = iterate([input_random_data])
            # update the input image in the direction that maximizes the loss
            input_random_data += grads_value * step
        plt.imshow(input_random_data.reshape(28, 28), cmap='viridis')
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
        plt.title('neuron '+str(j))
```

## 15.7 Variational Autoencoders (VAE)


- Encoder によって入力を $\mu$, $\sigma$ の2変数に落とし、$\mu+\epsilon\cdot\sigma$ ($\epsilon$はGaussノイズ) からサンプルした点をdecodeすることで、入力を復元するようなアーキテクチャを考える (variational autoencoder).


- コスト関数は MSE(input, output) に加え、latent の分布と Gauss 分布 $\mathcal{N}(0,1)$ との KL-divergence を考慮する.


- latent space を学習できるだけでなく、latent space をサンプリングすることで新しいインスタンスを生成することができる (generative model)


- ここでは、テキストと少し変えて MNIST を conv-net で学習する VAE を実装してみる [ref. Chollet "Deep Learning with Python"]

![title](png/15-11.png)

VAE のコスト関数の組み立てを考える (ref. Mehta et al., arXiv 2018)
![title](./png/mehta_VAE.png)

$p_\theta({\bf x}|{\bf z})$: latent space から新しい点を生成するためのモデルとする(decoder)。これは reconstruction error を lost function とすることで neural network として構成することができる。一方で、$p_\theta({\bf z}|{\bf x})$ (encoder) の構成が難しい。Baye's rule を考えると、
$$p_\theta({\bf z}|{\bf x}) = \frac{p({\bf z})p_{\theta}({\bf x}|{\bf z})}{p({\bf x})}$$
となるが、$p_\theta({\bf x}|{\bf z})$ が DNN なので解析的には解けない。そこで、別の encoder $q_\phi(\bf{z}|\bf{x})$ を考えて、それを $p_\theta({\bf x}|{\bf z})$ に近づけることを考える。2つの分布のKL-divergenceは次のように与えられる。
\begin{eqnarray}
D_{KL}(q_\phi(\bf{z}|\bf{x}) \| p_\theta(\bf{z}|\bf{x})) 
&=& \sum_{\bf z}q_\phi(\bf{z}|\bf{x}) \log \frac{q_\phi(\bf{z}|\bf{x})}{ p_\theta(\bf{z}|\bf{x})} \nonumber\\
&=& -\sum_{\bf z}q_\phi(\bf{z}|\bf{x}) \log \frac{p_\theta(\bf{z},\bf{x})}{p(\bf{x})}+ \sum_{\bf z}q_\phi(\bf{z}|\bf{x}) \log q_\phi(\bf{z}|\bf{x}) \nonumber\\
&=& \log p(\bf{x}) -\sum_{\bf z}q_\phi(\bf{z}|\bf{x}) \log p_\theta(\bf{z},\bf{x}) + \sum_{\bf z}q_\phi(\bf{z}|\bf{x}) \log q_\phi(\bf{z}|\bf{x}) \nonumber\\
&=& \log p(\bf{x}) -\sum_{\bf z}q_\phi(\bf{z}|\bf{x}) \log p_\theta(\bf{x}|\bf{z})+ \sum_{\bf z}q_\phi(\bf{z}|\bf{x}) \log\frac{ q_\phi(\bf{z}|\bf{x})}{p(\bf{z})} \nonumber\\
&=& \log p(\bf{x}) -\left\langle \log p_\theta(\bf{x}|\bf{z}) \right\rangle_{q_\phi(\bf{z}|\bf{x})}+ D_{KL}(q_\phi(\bf{z}|\bf{x})\|p(\bf{z})) \nonumber\\
\end{eqnarray} 


上の式を整理すると、


$$
\log p(\bf{x}) 
= D_{KL}(q_\phi(\bf{z}|\bf{x}) \| p_\theta(\bf{z}\|\bf{x})) 
+\left\langle \log p_\theta(\bf{x}|\bf{z}) \right\rangle_{q_\phi(\bf{z}|\bf{x})}
- D_{KL}(q_\phi(\bf{z}\|\bf{x})|p(\bf{z}))
$$


となる。$D_{KL}>0$なので、右辺第2,3項が log-likelihood $\log p(\bf{x}) $ の lower bound として働いていることがわかる。

VAEの文脈ではこれらの2項を合わせて Evidence Lower BOund (ELBO) と呼ばれる。

さらにこの式から、$D_{KL}(q_\phi(\bf{z}|\bf{x}) \| p_\theta(\bf{z}\|\bf{x})) $ を minimize したい場合、

- $\left\langle \log p_\theta(\bf{x}|\bf{z}) \right\rangle_{q_\phi(\bf{z}|\bf{x})}$ を上げる (reconstruction error を減らす)


- $D_{KL}(q_\phi(\bf{z}\|\bf{x})|p(\bf{z}))$ を小さくする (埋め込む latent space の分布を を $p(z)$ に近づける)


$p(z)$ の分布として multivariate Gaussian が選ばれることが多い (Gaussian VAE)

### Gaussian VAE の KL-divergence について

2つの正規分布 $\mathcal{N}(\mu_1,\sigma_1)$, $\mathcal{N}(\mu_2,\sigma_2)$ の間の Kullback-Leibler divergence は厳密に計算でき、次のように与えられる。


$$
D_{\rm KL}\left(\mathcal{N}(\mu_1,\sigma_1)\ ||\ \mathcal{N}(\mu_2,\sigma_2)\right) = 
\frac{1}{2}\left[\frac{\mu_1^2-2\mu_1\mu_2+\mu_2^2}{\sigma^2_2}
+\frac{\sigma^2_1}{\sigma_2^2}-1
-\ln{\frac{\sigma^2_1}{\sigma_2^2}}
\right]
$$


Gaussian VAEでは入力$X$に対して、$\mathcal{N}(\mu(X),\Sigma(X))$ を $\mathcal{N}(0,1)$ に近づけたいので、KL-divergence として、


$$
D_{\rm KL}\left(\mathcal{N}(\mu(X),\Sigma(X))\ ||\ \mathcal{N}(0,1)\right)=
\frac{1}{2}\left[\mu(X)^2+\Sigma(X)^2-1-\ln{\Sigma(X)^2} \right]
$$


を計算してあげればよい


```python
from keras import layers
# embedding datapoints into the latent space using conv-nets.(inference model)
img_shape = (28, 28, 1)
batch_size = 16
latent_dim = 2  # Dimensionality of the latent space

input_img = Input(shape=img_shape)
x = layers.Conv2D(32, 3, padding='same', activation='relu')(input_img)
x = layers.Conv2D(64, 3, padding='same', activation='relu', strides=(2, 2))(x)
x = layers.Conv2D(64, 3, padding='same', activation='relu')(x)
x = layers.Conv2D(64, 3, padding='same', activation='relu')(x)
shape_before_flattening = K.int_shape(x) # memorize the shape before flattening

x = layers.Flatten()(x)
x = layers.Dense(32, activation='relu')(x)

z_mean = layers.Dense(latent_dim)(x)
z_log_var = layers.Dense(latent_dim)(x)
```

```python
def sampling(args):
    """samples data points from the latent space
       given by a normal dist. N(z_mean,epsilon*exp(z_log_var))"""
    z_mean, z_log_var = args
    epsilon = K.random_normal(shape=(K.shape(z_mean)[0], latent_dim), mean=0., stddev=1.)
    return z_mean + K.exp(0.5*z_log_var) * epsilon

z = layers.Lambda(sampling)([z_mean, z_log_var])
```

```python
# decoding the sample point from the latent space (generative model)
decoder_input = layers.Input(K.int_shape(z)[1:])

# Upsample to the correct number of units
x = layers.Dense(np.prod(shape_before_flattening[1:]), activation='relu')(decoder_input)
x = layers.Reshape(shape_before_flattening[1:])(x)

x = layers.Conv2DTranspose(32, 3, padding='same', activation='relu', strides=(2, 2))(x)
x = layers.Conv2D(1, 3, padding='same', activation='sigmoid')(x)
decoder = Model(decoder_input, x)
z_decoded = decoder(z)
```

```python
# defining the loss function for Gaussian vae
class CustomVariationalLayer(keras.layers.Layer):

    def vae_loss(self, x, z_decoded):
        x = K.flatten(x)
        z_decoded = K.flatten(z_decoded)
        # normal loss function
        xent_loss = keras.metrics.binary_crossentropy(x, z_decoded)
        # KL-divergence
        kl_loss = -5e-4 * K.mean(1+z_log_var-K.square(z_mean)-K.exp(z_log_var), axis=-1)
        return K.mean(xent_loss + kl_loss)

    def call(self, inputs):
        x = inputs[0]
        z_decoded = inputs[1]
        loss = self.vae_loss(x, z_decoded)
        self.add_loss(loss, inputs=inputs)
        # We don't use this output.
        return x

y = CustomVariationalLayer()([input_img, z_decoded])
```

```python
# fitting model
vae = Model(input_img, y)
vae.compile(optimizer='adam', loss=None)
vae.summary()

(x_train, _), (x_test, y_test) = mnist.load_data()

x_train = x_train.astype('float32') / 255.
x_train = x_train.reshape(x_train.shape + (1,))
x_test = x_test.astype('float32') / 255.
x_test = x_test.reshape(x_test.shape + (1,))

vae.fit(x=x_train, y=None, shuffle=True, epochs=14,
        batch_size=batch_size, validation_data=(x_test, None))
# executed in 4min.1s
```

```python
# generating new data from the 2D latent space
from scipy.stats import norm

def plot_vae():
    n = 15  # figure with 15x15 digits
    digit_size = 28
    figure = np.zeros((digit_size * n, digit_size * n))
    # Linearly spaced coordinates on the unit square were transformed
    # through the inverse CDF (ppf) of the Gaussian
    # to produce values of the latent variables z,
    # since the prior of the latent space is Gaussian
    grid_x = np.linspace(-0.95, 0.95, n)  ###########################
    grid_y = np.linspace(-0.95, 0.95, n)  ###########################

    for i, yi in enumerate(grid_x):
        for j, xi in enumerate(grid_y):
            z_sample = np.array([[xi, yi]])
            # tile the same sample in order to match the size we can feed to the network
            z_sample = np.tile(z_sample, batch_size).reshape(batch_size, 2)
            x_decoded = decoder.predict(z_sample, batch_size=batch_size)
            digit = x_decoded[0].reshape(digit_size, digit_size)
            figure[i * digit_size: (i + 1) * digit_size,
                   j * digit_size: (j + 1) * digit_size] = digit

    plt.figure(figsize=(10, 10))
    ax = plt.axes()
    plt.imshow(figure, cmap='Greys_r')
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    plt.show()
```

latent space の点からサンプルして、新しいデータを生成するときは注意が必要。


いま latent space は $\mathcal{N}(0,1)$ に近くなっているので、単に線形にgridを取ると端の分解能が悪くなる(例えば上で生成された数字の中に1らしいものは見当たらない)。


したがって Gauss 分布の累積分布関数の逆関数を使うなどして、サンプル方法を工夫するとよい

```python
# generating new data from the 2D latent space
from scipy.stats import norm

def plot_vae():
    n = 15  # figure with 15x15 digits
    digit_size = 28
    figure = np.zeros((digit_size * n, digit_size * n))
    # Linearly spaced coordinates on the unit square were transformed
    # through the inverse CDF (ppf) of the Gaussian
    # to produce values of the latent variables z,
    # since the prior of the latent space is Gaussian
    grid_x = norm.ppf(np.linspace(0.05, 0.95, n))
    grid_y = norm.ppf(np.linspace(0.05, 0.95, n))

    for i, yi in enumerate(grid_x):
        for j, xi in enumerate(grid_y):
            z_sample = np.array([[xi, yi]])
            # tile the same sample in order to match the size we can feed to the network
            z_sample = np.tile(z_sample, batch_size).reshape(batch_size, 2)
            x_decoded = decoder.predict(z_sample, batch_size=batch_size)
            digit = x_decoded[0].reshape(digit_size, digit_size)
            figure[i * digit_size: (i + 1) * digit_size,
                   j * digit_size: (j + 1) * digit_size] = digit

    plt.figure(figsize=(10, 10))
    ax = plt.axes()
    plt.imshow(figure, cmap='Greys_r')
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    plt.show()
```
